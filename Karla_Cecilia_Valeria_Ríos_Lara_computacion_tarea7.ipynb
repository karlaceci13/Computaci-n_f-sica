Respuestas: 



*   **Método del despeje**

1.  Dividimos entre dos el polinomio.

 $10x^2 - 2x$ 

 $5x^2 - x$

2. Sacamos factor común.

  $x(5x - 1) = 0 $

3. Para que el polinomio sea igual a 0 es necesario que alguno de los dos elementos del producto obtenido sea igual a 0, así se obtiene un sistema de ecuaciones.
 
 $x=0$

 $5x-1=0$

4. Resultado del sistema de ecuaciones

   $x_1=0$

   $x_2=1/5$

*   **Método de la chicharronera**

1.  Dividimos entre dos el polinomio.

 $10x^2 - 2x$ 

 $5x^2 - x$

2. Identificamos los valores a,b,c respectivamente para hacer la fórmula general:

  $5x^2 - x + 0 = 0$

  $a= 5 , b=-1 , c= 0 $

3. Sustituimos en la fórmula y resolvemos: 

 $x= \frac{-b\pm \sqrt{b^2 - 4ac}}{2a}$ 

 $x = \frac{-(-1)\pm \sqrt{(-1^2) - 4(5)(0)}}{2(5)}$

 $ x = \frac{1 \pm \sqrt{1 - 0}}{10}$
 
 $ x = \frac{1 \pm {1}}{10}$

 $ x_1 = \frac{1 + {1}}{10} = \frac{0}{10}= 0 $ 

 $ x_2 = \frac{1 - {1}}{10}= \frac{2}{10}= \frac{1}{5}$ 

 $ x_1 = 0$

 $x_2 = \frac{1}{5}$

 


```python
#definimos la variable
a=0
def Comprobar(valor):
    x = valor

    y = x*(5*(x)-1)
    if (y==a):
        #Si "y" satisface la igualdad a cero, entonces se da el mensaje "Valor. Sí es raiz".
        print (x, "Sí es raíz")
    else:
        #Si "y" no satisface la igualdad a cero, entonces se da el mensaje "Valor. No es raíz".
        print (x, "No es raíz")

#comprobamos las raices
Comprobar(0)
Comprobar(1/5)
```

```python

def Chicharronera(a,b,c):
#usamos la formula general con el valor positivo y negativo
    x1 = (-b+((b**2-(4*a*c))**(1/2)))/(2*a)
    x2 = (-b-((b**2-(4*a*c))**(1/2)))/(2*a)
    #imprimimos las raices
    print("Las raicez son: {} y {}".format(x1,x2))

Chicharronera(10,-2,0)
```

Recuerda que el movimiento de caída libre simplificado cumple la ecuación: $y(t) = H + vt - \frac{1}{2}gt^2$.


```python
def Altura_MUA(t,v,y):
    g = 9.8 #valor de la gravedad
    H = y-(v*t)+((1/2)*g*(t**2)) #Usamos la formula correspondiente 

    print ("La altura en metros", H)
#damos el valor del tiempo en segundos para luego obtener la altura
tm = 0.05616667
s = tm*60
Altura_MUA(s,0,0)
```

```python
# Inidcamos que se eliga aleatoriamente un valor entre dos, una X cantidad de veces.
resultados_de_volados = np.random.randint(low=0, high=2, size=10000)
resultados_tabla = pd.DataFrame(resultados_de_volados, columns=["Resultados"])#obteniendo una tabla 
resultados_tabla.Resultados.replace({0:"Aguila", 1:"Sol"}, inplace=True)#resultados posibles del volado 
```

```python
fig = px.histogram(resultados_tabla, text_auto=True, labels={ #obtenemos una gráfica de los valores requeridos 
                        "value": "Resultados",
                        "count": "Frecuencia"}, template="plotly_white")

fig.update_layout(
    title="Simulación de tirar X volados",
    xaxis_title="Frecuencia",
    yaxis_title="Resultado"
    )
```
