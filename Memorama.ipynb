```python
import random
import time

# Símbolos eléctricos (10 pares)
simbolos = ["V", "A", "Ω", "⏚", "~", "⎓", "Hz", "W", "F", "H"]
cartas = simbolos * 2
random.shuffle(cartas)

# Estado del tablero
estado = ["X"] * 20
pares_encontrados = 0

def mostrar_tablero():
    for i in range(20):
        print(f"{estado[i]:>3}", end=" ")
        if (i + 1) % 5 == 0:
            print()
    print()

print("MEMORAMA: SÍMBOLOS ELÉCTRICOS")
print("Selecciona dos posiciones del 0 al 19\n")

while pares_encontrados < 10:
    mostrar_tablero()

    i1 = int(input("Primera carta: "))
    i2 = int(input("Segunda carta: "))

    if i1 == i2 or estado[i1] != "X" or estado[i2] != "X":
        print("Selección inválida\n")
        continue

    estado[i1] = cartas[i1]
    estado[i2] = cartas[i2]
    mostrar_tablero()

    if cartas[i1] == cartas[i2]:
        print("¡Par correcto!\n")
        pares_encontrados += 1
    else:
        print("No coinciden")
        time.sleep(1)
        estado[i1] = "X"
        estado[i2] = "X"

print("¡Juego terminado!")
```
